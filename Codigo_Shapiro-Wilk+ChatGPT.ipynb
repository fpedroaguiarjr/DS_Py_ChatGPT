```python
from scipy.stats import shapiro
```

```python
# AMOSTRA DE DADOS ALEATÓRIOS PARA EXEMPLO DE DISTRIBUIÇÃO NORMAL
data = np.random.normal(0,1,1000)
```

```python
# Para executar o teste Shapiro-Wilk
stat, p = shapiro(data)
```

```python
# Interpretando os resultados
alpha = 0.05
if p > alpha:
    print('Amostra parece ser normal(Não rejeitar H0)')
else:
    print('Amostra em desconformidade (Rejeitar H0)')
```

```python
# GERANDO OS GRÁFICOS
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
plt.hist(data, bins=30, density=True, alpha=0.6, color='g');
```

```python
# GERENDO A CURVA NORMAL
```

```python
mu, std = norm.fit(data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
plt.title("Histograma com curva normal")
plt.xlabel("Valores")
plt.ylabel("Densidade")
plt.show()
```
